# let's convert the data type of the "rate" column to float and remove the denominator.

```python
def handleRate(value):
    value=str(value).split('/')
    value=value[0];
    return float(value)

dataframe['rate']=dataframe['rate'].apply(handleRate)
print(dataframe.head())
```

# conclusion - There is no NULL value in the dataframe.

## Type of Resturant

```python
# Use a color palette for the bars with updated parameters
sns.set_theme(style="whitegrid")  # Optional: Set a theme
sns.countplot(
    x=dataframe['listed_in(type)'], 
    hue=dataframe['listed_in(type)'], 
    palette='Set2', 
    dodge=False, 
    legend=False
)
plt.xlabel("Type of restaurant", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.title("Distribution of Restaurant Types", fontsize=14)
plt.xticks(fontsize=10, rotation=0)  # No rotation for x-axis labels
plt.yticks(fontsize=10)

# Display the plot
plt.show()
```

# conclusion: The majority of the restaurants fall into the dining category.

# Dining resturants are preferred by a larger number of individuals.

```python
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes': grouped_data})
plt.plot(result, c="blue", marker="D")
plt.xlabel("Type of restaurant", c="red", size=20)
plt.ylabel("Votes", c="red", size=20)
```

# the majority of resturants received ratings

```python
plt.hist(dataframe['rate'], bins=5, color='skyblue')
plt.title("Ratings Distribution")
plt.show()


```

# whether online oreders receive higher ratings than offline orders.

```python
plt.figure(figsize = (6,6))
sns.boxplot(x = 'online_order', y = 'rate', data = dataframe, hue='online_order', palette=['lightblue', 'lightcoral'], legend=False)
```

```python
# Sample data
data = {
    'Type': ['Buffet', 'Cafes', 'Dining', 'Other'],
    'No': [3, 8, 77, 2],
    'Yes': [4, 15, 33, 6]
}

# Create DataFrame
df = pd.DataFrame(data)
df = df.set_index('Type')

# Plotting the heatmap
sns.heatmap(df, annot=True, cmap='YlGnBu', cbar_kws={'label': 'Rating'})
plt.title('Heatmap of Ratings by Type and Online Order', fontsize=16, color='blue')
plt.xlabel('Online Order', fontsize=14, color='green')
plt.ylabel('Listed In (Type)', fontsize=14, color='green')
plt.xticks(rotation=0, color='purple')
plt.yticks(rotation=0, color='purple')
plt.show()

```

# CONCLUSION: Dining resturants primarily accept offline orders, whereas cafes primarily recieve online orders. This suggests that cllients prefer to place orders in person at resturants,but prefer online ordering at cafes.

```python
#7)Which type of restaurant has the highest proportion
# Load the dataset
file_path = 'Zomato data .csv'  # Update with the correct path to your dataset file
zomato_data = pd.read_csv(file_path)

# Aggregating data for the pie chart: Distribution of restaurants by type
type_distribution = zomato_data['listed_in(type)'].value_counts()

# Plotting the pie chart
plt.figure(figsize=(6, 6))
type_distribution.plot.pie(
    autopct='%1.1f%%',
    startangle=140,
    colors=plt.cm.Paired.colors,
    explode=[0.05] * len(type_distribution),  # Adding spacing for better visibility
    shadow=True,
    labeldistance=1.05  # Adjusting label distance for proper alignment
)
plt.gca().set_aspect('equal')  # Ensures the pie chart is circular
plt.title('Distribution of Restaurants by Type', fontsize=14)
plt.ylabel('')  # Removing y-axis label for better aesthetics
plt.show()
```

# Conclusion:- The pie chart shows the distribution of restaurant types, highlighting the most common categories in the dataset. It reveals which restaurant types dominate and which are less prevalent, offering insights into market segmentation and trends within the dataset.

```python
#8) Which restaurant has the highest number of votes?
# Restaurant with the highest number of votes
top_voted_restaurant = dataframe.loc[dataframe['votes'].idxmax()]

# Visualization: Pie chart
other_votes = dataframe['votes'].sum() - top_voted_restaurant['votes']
labels = [top_voted_restaurant['name'], 'Other Restaurants']
sizes = [top_voted_restaurant['votes'], other_votes]
colors = ['#ff9999','#66b3ff']

plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title("Percentage of Votes for Top Voted Restaurant")
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()

print(f"Top Voted Restaurant: {top_voted_restaurant['name']} with {top_voted_restaurant['votes']} votes")
```

# Conclusion:-The pie chart shows the dominant share of votes held by the top-voted restaurant, with the remaining votes distributed among other restaurants. It highlights the popularity of the leading restaurant in the dataset.

```python
# 9) What is the average cost for two people across all restaurants?
# Convert to numeric if needed
if dataframe['approx_cost(for two people)'].dtype == 'object':
    dataframe['approx_cost(for two people)'] = dataframe['approx_cost(for two people)'].str.replace(',', '').astype(float)

# Average cost for two people across all restaurants
average_cost_for_two = dataframe['approx_cost(for two people)'].mean()

# Visualization
plt.figure(figsize=(8, 6))
sns.histplot(dataframe['approx_cost(for two people)'], bins=20, kde=True, color="purple")
plt.title("Distribution of Cost for Two People")
plt.xlabel("Cost for Two People (in currency)")
plt.ylabel("Number of Restaurants")
plt.show()

print(f"Average Cost for Two People: {average_cost_for_two:.2f}")
```

# Conclusion:The histogram shows the distribution of costs for two people, highlighting common price ranges and variability. The average cost gives an overall idea of typical restaurant pricing in the dataset.

```python
# 10)How many restaurants allow table booking ?
# Restaurants allowing table booking vs. those that donâ€™t
book_table_count = dataframe['book_table'].value_counts()

# Visualization
plt.figure(figsize=(8, 6))
sns.barplot(x=book_table_count.index, y=book_table_count.values, hue=book_table_count.index, palette="Greens_d", dodge=False, legend=False)
plt.title("Table Booking Availability")
plt.xlabel("Table Booking")
plt.ylabel("Number of Restaurants")
plt.show()

print(f"Table Booking Availability:\n{book_table_count}")
```

# Conclusion: The bar plot shows the number of restaurants offering table booking compared to those that do not, highlighting the availability of this feature in the dataset.
